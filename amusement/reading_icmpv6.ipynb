```python
from scapy.all import *

#affichage de toutes les trames
trames=rdpcap("icmpv6.pcapng")
trames.show()
```

```python
dic={128: "echo-request", 129: "echo-reply"}

numéro_de_trame=0
for trame in trames:
    try:
        #la trame contient du arp
        trames[13][IPv6][1].type        
        if trame[IPv6][1].type==128:
            print(f"====================")
            print(f"L'@MAC source est {trame[Ether].src}")
            print(f"L'@MAC destination est {trame[Ether].dst}")
            print(f"L'@IP source est {trame[IPv6].src}")
            print(f"L'@IP destination est {trame[IPv6].dst}")
            print(f"Requête de type echo-request")
        elif trame[IPv6][1].type==129:
            print(f"====================")
            print(f"L'@MAC source est {trame[Ether].src}")
            print(f"L'@MAC destination est {trame[Ether].dst}")
            print(f"L'@IP source est {trame[IPv6].src}")
            print(f"L'@IP destination est {trame[IPv6].dst}")
            print(f"Réponse de type echo-reply")
        numéro_de_trame+=1
    except:
        numéro_de_trame+=1
        continue
```
